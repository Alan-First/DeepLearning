# 文本分类任务

```python
from transformers import *
nlp_sentence_classif = pipeline("sentiment-analysis")
print(nlp_sentence_classif("I like this book!"))
```

# 特征提取

```python
# 返回文本处理后的特征，也就是可以作为词向量使用
import numpy as np
nlp_feature = pipeline('feature-extraction')
output = nlp_feature('Code Doctor Studio is a Chinese company based in BeiJing.')
print(np.array(output).shape)
```

```python
# 手动加载可以将文件夹'distilbert-base-cased'文件夹复制然后用跟文本分类任务一样的配置方式，
# 注意文件夹名是'./distilbert-base-cased/'
```

# 完形填空任务

```python
nlp_fill = pipeline('fill-mask')
print(nlp_fill.tokenizer.mask_token)
print(nlp_fill(f'Li Jinhong wrote many {nlp_fill.tokenizer.mask_token} about '+
    'artificial intelligence technology and helped many people.'))
```
